```python
import pandas as pd
import numpy as np
import json
import time
from tqdm import tqdm
from app_store_scraper import AppStore

def Extract_reviews(app_name: str, app_id: int): 
    camera_app = AppStore(country='kr', app_name=app_name, app_id=app_id)
    camera_app.review()

    if 'date' in camera_app.reviews:
        camera_app_reviews = pd.DataFrame(camera_app.reviews).sort_values('date', ascending=False).reset_index(drop=True)
    else:
        # 'date' 열이 없는 경우, 인덱스를 기준으로 정렬합니다.
        camera_app_reviews = pd.DataFrame(camera_app.reviews).sort_index(ascending=False).reset_index(drop=True)
    camera_app_reviews['camera_app'] = app_name
    return camera_app_reviews

camera_app_platform = {'concept': 6470380557}

app_store_reviews = pd.DataFrame()
for app_name, app_id in tqdm(camera_app_platform.items()):
    camera_app_reviews = Extract_reviews(app_name, app_id)
    app_store_reviews = pd.concat([app_store_reviews, camera_app_reviews], axis=0)

print('Done')
```

```python
app_store_reviews[app_store_reviews['camera_app'] == 'concept']
```

```python
app_store_reviews.to_excel('concept_review.xlsx')
```
